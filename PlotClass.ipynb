```python
import sys
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mpl
from matplotlib import cm
import matplotlib.font_manager as font_manager

sys.path.insert(0, '/Users/z5278074/Desktop/Class_Results')
```

```python
def MySquarePlot(xlab='',ylab='',\
                 lw=2,lfs=30,tfs=22,size_x=13,size_y=8,Grid=False):
    plt.rcParams['axes.linewidth'] = lw
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif',size=tfs)
    #mpl.rcParams['text.latex.preamble'] = [r'\usepackage{mathpazo}']

    fig = plt.figure(figsize=(size_x,size_y))
    ax = fig.add_subplot(111)

    ax.set_xlabel(xlab,fontsize=lfs)
    ax.set_ylabel(ylab,fontsize=lfs)

    ax.tick_params(which='major',direction='in',width=0.8,length=8,right=True,top=True,pad=7)
    ax.tick_params(which='minor',direction='in',width=0.8,length=8,right=True,top=True)
    if Grid:
        ax.grid()
    return fig,ax
```

## Load files 

```python
# New smooth functions files 

Nfiles = 23

f = [ np.loadtxt(f"f{i}.dat", usecols=(0, 1), unpack=True) for i in range(Nfiles+1)]
lcdm = np.loadtxt("LCDM.dat", usecols=(0, 1), unpack=True)
```

# Scenario A example

```python
fig,ax = MySquarePlot(r'$\ell$',r'$C_{\ell}/C^{\rm \Lambda CDM}_{\ell}$')

x = np.arange(2,2500,10)
col = cm.summer(np.linspace(0,0.8,6))

ax.plot(f[17][0],f[1][1]/f[17][1],label=r'$Y=10 ~s^{-1}$',c=col[0],lw=2.5)
ax.plot(f[17][0],f[2][1]/f[17][1],label=r'$Y=10^2 ~s^{-1}$',c=col[1],lw=2.5)
ax.plot(f[17][0],f[3][1]/f[17][1],label=r'$Y=10^3 ~s^{-1}$',c=col[2],lw=2.5)
ax.plot(f[17][0],f[4][1]/f[17][1],label=r'$Y=10^4 ~s^{-1}$',c=col[3],lw=2.5)
ax.plot(f[17][0],f[5][1]/f[17][1],label=r'$Y=10^5 ~s^{-1}$',c=col[4],lw=2.5)
ax.plot(f[17][0],f[6][1]/f[17][1],label=r'$Y=10^8 ~s^{-1}$',c=col[5],lw=2.5)


ax.plot(x,x/x,ls='--',c='k')
ax.set_title(r'${\rm Scenario~A},~m_{\nu H}=0.05~\rm{eV}$')

ax.set_xlim(2,2500)
ax.legend(frameon=False)
plt.show()
```

# Varying Y

```python
fig,ax = MySquarePlot(r'$\ell$',r'$C_{\ell}/C^{\rm \Lambda CDM}_{\ell}-1$',size_x=13,size_y=8)

lw = 3

x = np.arange(2,4000,1)
ax.plot(x,x/x-1,ls='--',c='k')
col = cm.summer(np.linspace(0,0.7,3))

ax.semilogx(f[17][0],f[3][1]/f[17][1]-1,label=r'${\rm Scenario~A,~}Y=10^3 ~s^{-1}$',c=col[0],lw=lw,ls='--')
ax.semilogx(f[17][0],f[4][1]/f[17][1]-1,label=r'${\rm Scenario~A,~}Y=10^4 ~s^{-1}$',c=col[1],lw=lw,ls='--')
ax.semilogx(f[17][0],f[5][1]/f[17][1]-1,label=r'${\rm Scenario~A,~}Y=10^5 ~s^{-1}$',c=col[2],lw=lw,ls='--')

ax.semilogx(lcdm[0],f[9][1]/lcdm[1]-1,c=col[0],lw=lw,label=r'${\rm Scenario~B,~}Y=10^3 ~s^{-1}$')
ax.semilogx(lcdm[0],f[10][1]/lcdm[1]-1,c=col[1],lw=lw,label=r'${\rm Scenario~B,~}Y=10^4 ~s^{-1}$')
ax.semilogx(lcdm[0],f[11][1]/lcdm[1]-1,c=col[2],lw=lw,label=r'${\rm Scenario~B,~}Y=10^5 ~s^{-1}$')

ax.set_xlim(2,2500)
ax.set_ylim(-0.0075,0.078)
ax.set_title(r'$m_{\nu H}=0.05~\rm{eV}$')
ax.legend(frameon=False,loc=2,ncol=2,prop={'size': 19})

plt.show()

fig.savefig('/Users/z5278074/Desktop/TTplot_Y.pdf',bbox_inches='tight')
```
